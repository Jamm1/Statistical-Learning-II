```python
import pandas as pd
import numpy as np
df = pd.read_csv('supermarket_sales.csv')
df.head()
```

```python

df.isnull().values.any()
```

```python
df = df.drop(['Invoice ID', 'Date', 'Time',], axis = 1)
df.head()
```

```python
ohe = pd.get_dummies(df['City'], drop_first=True)
df.drop('City', axis=1, inplace=True)
df = df.join(ohe)
df.head()
```

```python
ohe = pd.get_dummies(df['Gender'], drop_first=True)
df.drop('Gender', axis=1, inplace=True)
df = df.join(ohe)
df.head()
```

```python
ohe = pd.get_dummies(df['Branch'], drop_first=True)
df.drop('Branch', axis=1, inplace=True)
df = df.join(ohe)
df.head()
```

```python
ohe = pd.get_dummies(df['Product line'], drop_first=True)
df.drop('Product line', axis=1, inplace=True)
df = df.join(ohe)
df.head()
```

```python
ohe = pd.get_dummies(df['Customer type'], drop_first=True)
df.drop('Customer type', axis=1, inplace=True)
df = df.join(ohe)
df.head()
```

```python
ohe = pd.get_dummies(df['Payment'], drop_first=True)
df.drop('Payment', axis=1, inplace=True)
df = df.join(ohe)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
```

```python
X = df.iloc[:, df.columns != 'Quantity']
y = df.loc[:, 'Quantity']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
X_test.shape[0]
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
entradas = X_train.shape[1]
lr = 0.01
batch_sz=10

regressor = Sequential()
regressor.add(Dense(input_dim=entradas, units=7, kernel_initializer='uniform', activation='relu'))
regressor.add(Dense(units=7, kernel_initializer='uniform', activation='relu'))
regressor.add(Dense(units=3, kernel_initializer='uniform', activation='relu'))
regressor.add(Dense(units=1, kernel_initializer='uniform', activation='relu'))
regressor.compile(optimizer=Adam(lr), loss='mean_squared_error', metrics=['mean_squared_error'])
resultado1 = regressor.fit(X_train, y_train, batch_size=batch_sz, validation_split=0.2, verbose=1, epochs=100)
resultado1.history

```

```python
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasRegressor
import time
```

```python
parametros = {
    'batch_size': [5, 11,15],
    'nb_epoch':[100, 150],
    'optimizer': ['adam',]
}
```

```python
def reg_gs_v1(optimizer):
  regressor = Sequential()
  regressor.add(Dense(input_dim=entradas, units=3, kernel_initializer='uniform', activation='relu'))
  regressor.add(Dense(units=8, kernel_initializer='uniform', activation='relu'))
  regressor.add(Dense(units=2, kernel_initializer='uniform', activation='relu'))
  regressor.add(Dense(units=1, kernel_initializer='uniform', activation='relu'))
  regressor.compile(optimizer=optimizer, loss='mean_squared_error', metrics=['mean_squared_error'])
  result1 = regressor.fit(X_train, y_train, batch_size=batch_sz, validation_split=0.2, verbose=1, epochs=50)
  return regressor
```

```python
regresor_grid_search = KerasRegressor(build_fn=reg_gs_v1)
```

```python
grid_out = GridSearchCV(estimator=regresor_grid_search,
                        param_grid=parametros,
                        cv=10,
                        scoring='neg_mean_squared_error')
```

```python
train_gs_out = grid_out.fit(X_train, y_train, verbose=1)
```
